```python
import re
```

```python
def sanitize_col_name(col_name: str) -> str:
    col_name = col_name.lower()
    col_name = re.sub(r'[áàãâä]', 'a', col_name)
    col_name = re.sub(r'[éèêë]', 'e', col_name)
    col_name = re.sub(r'[íìîï]', 'i', col_name)
    col_name = re.sub(r'[óòõôö]', 'o', col_name)
    col_name = re.sub(r'[úùûü]', 'u', col_name)
    col_name = re.sub(r'[ç]', 'c', col_name)
    col_name = re.sub(r'[^a-z0-9]+', '_', col_name)
    col_name = col_name.strip('_')
    return col_name
```
