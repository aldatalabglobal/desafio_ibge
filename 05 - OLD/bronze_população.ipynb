```python
# Ferramentas para ingestão da API 
import requests # Biblioteca padrão para fazer requisições HTTP
from requests.adapters import HTTPAdapter #Se a API do IBGE falhar, o código tentará novamente automaticamente
from urllib3.util import Retry #Evita que o pipeline falhe por instabilidades temporárias da rede ou da API do IBGE
import json # Analisa (parsear) a resposta da API do IBGEe a transforma em um objeto Python utilizável

# Ferramentas para processamento de dados 
import pandas as pds # Biblioteca padrão Python para manipulação de dados em memória
from pyspark import pandas as pd # A versão do Pandas otimizada para trabalhar com DataFrames PySpark
import numpy as np # Biblioteca padrão Python para computação numérica

# Ferramentas de Sistema e Delta Lake
import os # Permite interagir com o sistema operacional
import datetime # Usado para obter a data e hora atual
from delta.tables import * # Permite interagir diretamente com as funcionalidades avançadas do Delta Lake
```

```python
url_api_populacao= 'https://apisidra.ibge.gov.br/values/t/793/n3/all/v/93/p/all/f/u'

retry_strategy = Retry(
    total=4,
    status_forcelist=[429, 500, 502, 503, 504],
)
adapter = HTTPAdapter (max_retries=retry_strategy)
session = requests.Session() 
session.mount(url_api_populacao, adapter)
response = session.get(url_api_populacao)
```
